Analisis Kilometraje vs precio promedio ( por marca de auto)

```python
# Cargar el archivo CSV
df = pd.read_csv("cars (1).csv")

# Filtrar datos válidos
df_filtered = df[(df['price_usd'] > 1000) & 
                 (df['odometer_value'] > 0) & 
                 (df['odometer_value'] < 500000)]

# Agrupar por marca y calcular estadísticas
brand_stats = df_filtered.groupby('manufacturer_name').agg({
    'price_usd': 'mean',
    'odometer_value': 'mean',
    'year_produced': 'mean',
    'manufacturer_name': 'count'
}).rename(columns={
    'price_usd': 'avg_price',
    'odometer_value': 'avg_km',
    'year_produced': 'avg_year',
    'manufacturer_name': 'num_cars'
}).sort_values(by='avg_price', ascending=False)

# Seleccionar las 10 marcas más caras
top_brands = brand_stats.head(10).reset_index()

# Visualización
sns.set(style="whitegrid")
plt.figure(figsize=(16, 8))

# Precio vs Kilometraje
plt.subplot(1, 2, 1)
sns.scatterplot(data=top_brands, x='avg_km', y='avg_price', hue='manufacturer_name', s=100)
plt.title('Precio vs Kilometraje Promedio (Top 10 marcas)')
plt.xlabel('Kilometraje Promedio')
plt.ylabel('Precio Promedio (USD)')

# Precio vs Año
plt.subplot(1, 2, 2)
sns.scatterplot(data=top_brands, x='avg_year', y='avg_price', hue='manufacturer_name', s=100)
plt.title('Precio vs Año de Producción Promedio (Top 10 marcas)')
plt.xlabel('Año Promedio')
plt.ylabel('Precio Promedio (USD)')

plt.tight_layout()
plt.show()
```

Análisis de tipo de transmisión

```python
# Filtrar datos válidos
df_filtered = df[(df['price_usd'] > 1000) & (df['price_usd'] < 100000)]

# Agrupar por tipo de transmisión
transmission_stats = df_filtered.groupby('transmission').agg({
    'price_usd': 'mean',
    'transmission': 'count'
}).rename(columns={
    'price_usd': 'avg_price',
    'transmission': 'count'
}).reset_index()

# Configurar estilo
sns.set(style="whitegrid")

# Crear figura y ejes
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico 1: Cantidad por tipo de transmisión
sns.barplot(data=transmission_stats, x='transmission', y='count', ax=axes[0], palette='Blues_d')
axes[0].set_title('Cantidad de Autos por Tipo de Transmisión')
axes[0].set_ylabel('Cantidad')
axes[0].set_xlabel('Transmisión')

# Gráfico 2: Precio promedio por tipo de transmisión
sns.barplot(data=transmission_stats, x='transmission', y='avg_price', ax=axes[1], palette='Greens_d')
axes[1].set_title('Precio Promedio por Tipo de Transmisión')
axes[1].set_ylabel('Precio Promedio (USD)')
axes[1].set_xlabel('Transmisión')

# Ajustar y mostrar
plt.tight_layout()
plt.show()
```

Análisis de Año vs Precio y Cantidad

```python
# Filtrar datos válidos
df_filtered = df[(df['price_usd'] > 1000) & 
                 (df['price_usd'] < 100000) & 
                 (df['year_produced'] >= 1980) & 
                 (df['year_produced'] <= 2022)]

# Agrupar por año y calcular precio promedio y cantidad de autos
year_stats = df_filtered.groupby('year_produced').agg({
    'price_usd': 'mean',
    'manufacturer_name': 'count'
}).rename(columns={
    'price_usd': 'avg_price',
    'manufacturer_name': 'num_cars'
}).reset_index()

# Configurar estilo
sns.set(style="whitegrid")
plt.figure(figsize=(16, 8))

# Crear gráfico de línea doble (dos ejes)
fig, ax1 = plt.subplots(figsize=(16, 8))

# Primer eje: Precio promedio
sns.lineplot(data=year_stats, x='year_produced', y='avg_price', ax=ax1, color='blue', label='Precio Promedio')
ax1.set_ylabel('Precio Promedio (USD)', color='blue')
ax1.set_xlabel('Año de Fabricación')
ax1.tick_params(axis='y', labelcolor='blue')

# Segundo eje: Cantidad de autos
ax2 = ax1.twinx()
sns.lineplot(data=year_stats, x='year_produced', y='num_cars', ax=ax2, color='green', label='Cantidad de Autos')
ax2.set_ylabel('Cantidad de Autos', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Título y leyenda
plt.title('Relación entre Año de Fabricación, Precio Promedio y Disponibilidad')
fig.tight_layout()
plt.show()
```

Análisis de Eficiencia por Tipo de Combustible

```python
# Cargar el dataset
df = pd.read_csv("cars (1).csv")

# Limpiar y filtrar datos
df_filtered = df[(df['odometer_value'] > 0) & 
                 (df['engine_capacity'] > 0) & 
                 (df['price_usd'] > 1000) & 
                 (df['price_usd'] < 100000)]

# Calcular eficiencia aproximada: km por litro de capacidad nominal
df_filtered['efficiency_kpl'] = df_filtered['odometer_value'] / df_filtered['engine_capacity']

# Agrupar por tipo de combustible y calcular eficiencia promedio
fuel_efficiency = df_filtered.groupby('engine_fuel')['efficiency_kpl'].mean().reset_index()
fuel_efficiency = fuel_efficiency.sort_values(by='efficiency_kpl', ascending=False)

# Visualizar
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(data=fuel_efficiency, x='efficiency_kpl', y='engine_fuel', palette='viridis')

plt.title('Eficiencia Promedio por Tipo de Combustible (km por litro de motor)')
plt.xlabel('Eficiencia Aproximada (km / litro motor)')
plt.ylabel('Tipo de Combustible')

plt.tight_layout()
plt.show()
```

heatmap tipo “semáforo” para identificar qué autos son buena o mala inversión visualmente.

```python
# Cargar dataset
df = pd.read_csv("cars (1).csv")

# Filtrar datos relevantes
df = df[(df['price_usd'] > 1000) & (df['price_usd'] < 100000) & 
        (df['year_produced'] > 1990) & (df['odometer_value'] > 0)]

# Crear columna combinada de marca y modelo
df['full_model'] = df['manufacturer_name'] + ' ' + df['model_name']

# Agrupar por modelo completo
grouped = df.groupby('full_model').agg({
    'price_usd': 'mean',
    'odometer_value': 'mean',
    'year_produced': 'mean',
    'full_model': 'count'
}).rename(columns={
    'price_usd': 'avg_price',
    'odometer_value': 'avg_km',
    'year_produced': 'avg_year',
    'full_model': 'num_units'
}).reset_index()

# Normalizar (escalar entre 0 y 1)
grouped['price_score'] = 1 - (grouped['avg_price'] - grouped['avg_price'].min()) / (grouped['avg_price'].max() - grouped['avg_price'].min())
grouped['km_score'] = 1 - (grouped['avg_km'] - grouped['avg_km'].min()) / (grouped['avg_km'].max() - grouped['avg_km'].min())
grouped['year_score'] = (grouped['avg_year'] - grouped['avg_year'].min()) / (grouped['avg_year'].max() - grouped['avg_year'].min())
grouped['demand_score'] = (grouped['num_units'] - grouped['num_units'].min()) / (grouped['num_units'].max() - grouped['num_units'].min())

# Calcular puntuación final de inversión
grouped['investment_score'] = (grouped['price_score'] + grouped['km_score'] + grouped['year_score'] + grouped['demand_score']) / 4

# Seleccionar top 15 modelos más rentables
top_models = grouped.sort_values(by='investment_score', ascending=False).head(15)

# Heatmap
heat_data = top_models[['price_score', 'km_score', 'year_score', 'demand_score', 'investment_score']]
heat_data.index = top_models['full_model']

plt.figure(figsize=(14, 8))
sns.heatmap(heat_data, annot=True, cmap="RdYlGn", linewidths=.5)
plt.title('Puntuación de Inversión por Modelo (0 = malo, 1 = excelente)')
plt.tight_layout()
plt.show()
```
